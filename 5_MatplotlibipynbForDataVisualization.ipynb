```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(0, 5, 11)
```

```python
y = x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x, y, 'r')
```

```python
plt.xlabel('X axis Title Here')
plt.ylabel('Y axis Title Here')
plt.title('String Title Here')
plt.show()
```

```python
#Subplots
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r--')
plt.subplot(1, 2, 2)
plt.plot(y, x, 'g*-')
```

```python
#Second Method that is object oriented 
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y, 'b')
axes.set_xlabel('Set X Label')
axes.set_ylabel('Set Y label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x, y, 'b')
axes1.set_xlabel('X_label_axes1')
axes1.set_ylabel('Y_label_axes1')
axes1.set_title('Axes 1 title')

axes2.plot(x, y, 'b')
axes2.set_xlabel('X_label_axes2')
axes2.set_ylabel('Y_label_axes2')
axes2.set_title('Axes 2 title')

```

```python
#Subplots
fig, axes = plt.subplots()

axes.plot(x, y, 'r')
axes.set_xlabel('s')
axes.set_ylabel("Title")
```

```python
fig, axes = plt.subplots(nrows = 1, ncols=2)
```

```python
axes
```

```python
for ax in axes:
  ax.plot(x, y, 'b')
  ax.set_xlabel('x')
  ax.set_ylabel('y')
  ax.set_title('title')

fig
```

```python
#recommended
fig, axes = plt.subplots(nrows = 1, ncols=2)
for ax in axes:
  ax.plot(x, y, 'b')
  ax.set_xlabel('x')
  ax.set_ylabel('y')
  ax.set_title('title')

fig
plt.tight_layout()
```

```python
#legends

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label="x**2")
ax.plot(x, x**3, label="x**3")
#location of Legends
ax.legend(loc=2)
```

```python
#colors

fig, ax = plt.subplots()
ax.plot(x, x**2, 'b--')
ax.plot(x, x**3, 'g--')
```

```python
fig, ax = plt.subplots()
ax.plot(x, x+1, color='blue', alpha=0.5)
ax.plot(x, x+2, color='#880088')
ax.plot(x, x+3, color='#FF8c00')
```

```python
#Line and Marker Style
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(x, x+1, color='blue', linewidth=2)
ax.plot(x, x+2, color='blue', linewidth=0.5)
ax.plot(x, x+3, color='blue', linewidth=0.8)
ax.plot(x, x+4, color='blue', linewidth=1)
#linewidth=lu shortcut
ax.plot(x, x+5, color='red', lw=3,ls='-')
ax.plot(x, x+6, color='red', lw=3, ls='-.')
ax.plot(x, x+7, color='red', linewidth=4, ls=':')

line, = ax.plot(x, x+8, color="black", lw=1.5)
line.set_dashes([5, 10, 15, 10])

ax.plot(x, x+9, color='blue', lw=3, ls='-', marker='o', markersize=10, markerfacecolor='red' , markeredgewidth=3, markeredgecolor='green')
```

```python
# Axis Appearance
fig,  axes = plt.subplots(1, 3, figsize=(12, 4))

axes[0].plot(x, x**2, x, x**3)
axes[0].set_title('Default axes ranges')

axes[1].plot(x, x**2, x, x**3)
axes[1].axis('tight')
axes[1].set_title('tight axes')

axes[2].plot(x, x**2, x, x**3)
axes[2].set_ylim([0, 60])
axes[2].set_xlim([2, 5])
axes[2].set_title('custom axes range')
```
